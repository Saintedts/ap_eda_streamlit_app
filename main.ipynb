```python
import pandas as pd 
import numpy as np 
```

```python
d_work = pd.read_csv('datasets/D_work.csv')
d_close_loan = pd.read_csv('datasets/D_close_loan.csv')
d_job = pd.read_csv('datasets/D_job.csv')
d_last_credit = pd.read_csv('datasets/D_last_credit.csv')
d_loan = pd.read_csv('datasets/D_loan.csv')
d_pens = pd.read_csv('datasets/D_pens.csv')
d_salary = pd.read_csv('datasets/D_salary.csv')
d_target = pd.read_csv('datasets/D_target.csv')
d_work = pd.read_csv('datasets/D_work.csv')
d_clients = pd.read_csv('datasets/D_clients.csv')
```

```python
main_df = d_target.copy()
main_df.head(10)
```

```python
main_df = main_df.merge(d_clients, how='left', left_on='ID_CLIENT', right_on='ID').drop(columns='ID')
main_df = main_df.merge(d_salary, how='left', left_on='ID_CLIENT', right_on='ID_CLIENT' )
main_df.head(10)
```

```python
sub_query = d_loan.merge(d_close_loan, how='left', left_on='ID_LOAN', right_on='ID_LOAN')
sub_query = sub_query.groupby(by = 'ID_CLIENT', as_index=False).count()
sub_query.columns = ['ID_CLIENT', 'LOAN_NUM_TOTAL', 'LOAN_NUM_CLOSED']
sub_query.head(10)
```

```python
main_df = main_df.merge(sub_query, how='left', left_on='ID_CLIENT', right_on='ID_CLIENT')
main_df.head(15)
```

```python
main_df = main_df.drop_duplicates(subset='ID_CLIENT', keep='last')
main_df.to_csv('datasets/main.csv', index=False)
```
